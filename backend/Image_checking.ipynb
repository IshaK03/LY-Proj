```python
import os
from PIL import Image

def find_corrupt_images(folder_path):

    corrupt_images = []

    for root, _, files in os.walk(folder_path):
        for fname in files:
            file_path = os.path.join(root, fname)
            print(f"Checking: {file_path}")
            try:
                with Image.open(file_path) as img:
                    img.verify()
            except Exception as e:
                print(f"Corrupt image found: {file_path}. Error: {e}")
                corrupt_images.append(file_path)

    return corrupt_images


# Provide your directory here
folder = "/content/drive/MyDrive/IAM_DATASET/iam_words/words/"

corrupt = find_corrupt_images(folder)
print("\nSummary:")
print("Corrupt files:")
for f in corrupt:
    print(f)
```
